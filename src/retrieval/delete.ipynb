```python
import chromadb
from chromadb.config import Settings
from chromadb.utils import embedding_functions
from dotenv import load_dotenv
import os
from sentence_transformers import SentenceTransformer
from chromadb import Documents, EmbeddingFunction, Embeddings
load_dotenv()
```

```python

# hf_embed = embedding_functions.HuggingFaceEmbeddingFunction(
#     model_name="Qwen/Qwen3-Embedding-0.6B",
#     api_key=os.getenv("HUGGINGFACE_API_KEY")
# )
model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")


class MyEmbeddingFunction(EmbeddingFunction):
    def __call__(self, input: Documents) -> Embeddings:
        # embed the documents somehow
        return model.encode(input, normalize_embeddings=True).tolist()
def st_embed(texts: list[str]) -> list[list[float]]:
    # Returns a 2D list of floats
    return model.encode(texts, normalize_embeddings=True).tolist()

client = chromadb.Client(Settings(
    persist_directory="./chroma_db"))

collection = client.create_collection(
    name="my_hf_collection_3212121221113",
    embedding_function=MyEmbeddingFunction()
)
```

```python
collection.add(
    documents=[
        "Machine learning enables computers to learn vector from data.",
        "ChromaDB is a vector database for AI applications."
    ],
    # metadatas=[{"topic": "ml"}, {"topic": "vector-db"}],
    ids=["1", "2"]
)
```

```python
results = collection.query(
    query_texts=["vector"],
    n_results=2
)
print(results)
```
