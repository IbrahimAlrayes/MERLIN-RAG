```python
import weaviate
from weaviate.classes.config import Configure, Property, DataType

# Connection settings
WEAVIATE_IP = "128.2.205.27"
HTTP_PORT = 8080
GRPC_PORT = 50051

# Connect to Weaviate
client = weaviate.connect_to_custom(
    http_host=WEAVIATE_IP,
    http_port=HTTP_PORT,
    http_secure=False,
    grpc_host=WEAVIATE_IP,
    grpc_port=GRPC_PORT,
    grpc_secure=False,
)

try:
    # Define collection name
    collection_name = "TestArticles"
    
    # Delete collection if it exists (for clean testing)
    if client.collections.exists(collection_name):
        client.collections.delete(collection_name)
        print(f"Deleted existing '{collection_name}' collection")
    
    # Create a new collection with vectorization
    # Using text2vec-transformers as the vectorizer
    collection = client.collections.create(
        name=collection_name,
        vectorizer_config=Configure.Vectorizer.text2vec_transformers(),
        properties=[
            Property(
                name="title",
                data_type=DataType.TEXT,
                vectorize_property_name=True,  # Include property name in vectorization
                skip_vectorization=False  # This property WILL be vectorized
            ),
            Property(
                name="content",
                data_type=DataType.TEXT,
                vectorize_property_name=False,
                skip_vectorization=False  # This property WILL be vectorized
            ),
            Property(
                name="category",
                data_type=DataType.TEXT,
                vectorize_property_name=False,
                skip_vectorization=True  # This property will NOT be vectorized
            ),
            Property(
                name="views",
                data_type=DataType.INT,
                skip_vectorization=True  # Numbers are not vectorized
            )
        ]
    )
    
    print(f"\nтЬУ Created collection '{collection_name}' with vectorization enabled")
    
    # Insert test data
    test_data = [
        {
            "title": "Introduction to Machine Learning",
            "content": "Machine learning is a subset of artificial intelligence that enables systems to learn and improve from experience.",
            "category": "AI",
            "views": 1500
        },
        {
            "title": "Python Programming Basics",
            "content": "Python is a versatile programming language known for its simplicity and readability, perfect for beginners.",
            "category": "Programming",
            "views": 2300
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Deep Learning Neural Networks",
            "content": "Neural networks are computing systems inspired by biological neural networks that constitute animal brains.",
            "category": "AI",
            "views": 1800
        },
        {
            "title": "Web Development with Flask",
            "content": "Flask is a lightweight WSGI web application framework in Python designed to make getting started quick and easy.",
            "category": "Web",
            "views": 950
        }
    ]
    
    # Insert objects
    articles = client.collections.get(collection_name)
    for i, data in enumerate(test_data):
        uuid = articles.data.insert(data)
        print(f"  Inserted object {i+1}: {data['title']} (UUID: {uuid})")
    
    print(f"\nтЬУ Inserted {len(test_data)} objects")
    
    # Verify vectorization configuration
    print(f"\n=== Collection Configuration ===")
    config = articles.config.get()
    print(f"Vectorizer: {config.vectorizer}")
    print("\nProperties:")
    for prop in config.properties:
        vectorized = not getattr(prop, "skip_vectorization", True)
        print(f"  - {prop.name} ({prop.data_type}): Vectorized = {vectorized}")
    
    # Perform a vector search to test
    print(f"\n=== Testing Vector Search ===")
    query_text = "artificial intelligence and neural networks"
    print(f"Query: '{query_text}'")
    
    response = articles.query.near_text(
        query=query_text,
        limit=3
    )
    
    print(f"\nTop {len(response.objects)} results:")
    for i, obj in enumerate(response.objects):
        print(f"\n{i+1}. {obj.properties['title']}")
        print(f"   Category: {obj.properties['category']}")
        print(f"   Content: {obj.properties['content'][:80]}...")
        print(f"   Distance: {obj.metadata.distance:.4f}" if obj.metadata.distance else "")
    
    # List all collections
    print(f"\n=== All Collections ===")
    collections = client.collections.list_all()
    for name in collections:
        print(f"\n{name}")
        coll_config = client.collections.get(name).config.get()
        vectorized = [p.name for p in coll_config.properties if not getattr(p, "skip_vectorization", True)]
        print(f"  Vectorized properties: {vectorized or 'None'}")

finally:
    # Close connection
    client.close()
    print("\nтЬУ Connection closed")
```

```python
import weaviate
from weaviate.classes.config import Configure, Property, DataType
import uuid

WEAVIATE_IP = "128.2.205.27"
HTTP_PORT = 8080
GRPC_PORT = 50051

client = weaviate.connect_to_custom(
    http_host=WEAVIATE_IP,
    http_port=HTTP_PORT,
    http_secure=False,
    grpc_host=WEAVIATE_IP,
    grpc_port=GRPC_PORT,
    grpc_secure=False,
)

try:
    print("=" * 70)
    print("DEBUGGING: Why does TestArticles work but New collection fails?")
    print("=" * 70)
    
    # Check both collections
    for coll_name in ["TestArticles", "New"]:
        print(f"\n{'='*70}")
        print(f"Checking: {coll_name}")
        print('='*70)
        
        if not client.collections.exists(coll_name):
            print(f"тЭМ Collection '{coll_name}' does not exist")
            continue
            
        col = client.collections.get(coll_name)
        config = col.config.get()
        
        print(f"\nЁЯУЛ Vectorizer: {config.vectorizer}")
        print(f"ЁЯУЛ Vectorizer Config: {config.vectorizer_config}")
        
        print(f"\nЁЯУЭ Properties:")
        for prop in config.properties:
            skip = getattr(prop, "skip_vectorization", True)
            print(f"   {prop.name} ({prop.data_type})")
            print(f"      skip_vectorization: {skip}")
            print(f"      vectorize_property_name: {getattr(prop, 'vectorize_property_name', 'N/A')}")
    
    # Now let's try to insert into the 'New' collection with the SAME data structure
    # that works for TestArticles
    print(f"\n{'='*70}")
    print("TEST: Insert simple data into 'New' collection")
    print('='*70)
    
    if client.collections.exists("New"):
        new_col = client.collections.get("New")
        
        # Try inserting minimal test data (similar to TestArticles)
        test_props = {
            "title": "Test Article",
            "summary": "This is a test summary",
            "content": "This is test content",
            "lang": "en",
            "qid": "Q12345",
            "wikipedia_pageviews_90d": 100,
            "wikipedia_backlinks": 10,
            "article_size_bytes": 1000,
            "reference_count": 5,
            "revision_count": 20,
            "unique_editors": 3,
            "category_count": 2,
            "image_count": 1,
            "external_links": 5,
            "wikidata_incoming_links": 8,
            "wikidata_outgoing_links": 15,
            "language_editions": 25,
            "statement_count": 30,
            "has_wikidata_image": True,
            "qualifier_count": 10,
            "entity_age_days": 365,
            "entity_age_years": 1.0,
        }
        
        try:
            test_uuid = new_col.data.insert(properties=test_props)
            print(f"тЬЕ SUCCESS! Inserted test object with UUID: {test_uuid}")
            print("\nThis means the 'New' collection CAN work with proper data!")
        except Exception as e:
            print(f"тЭМ FAILED to insert:")
            print(f"   Error: {e}")
            print("\nLet's try with ONLY the required fields:")
            
            # Try with minimal fields
            minimal_props = {
                "title": "Minimal Test",
                "summary": "Test",
                "content": "Test content",
                "lang": "en",
                "qid": "Q99999"
            }
            
            try:
                minimal_uuid = new_col.data.insert(properties=minimal_props)
                print(f"тЬЕ Minimal insert worked! UUID: {minimal_uuid}")
            except Exception as e2:
                print(f"тЭМ Even minimal insert failed: {e2}")
    
    # Check what properties are actually required vs optional
    print(f"\n{'='*70}")
    print("PROPERTY ANALYSIS")
    print('='*70)
    
    if client.collections.exists("New"):
        config = client.collections.get("New").config.get()
        print("\nProperty definitions in 'New' collection:")
        for prop in config.properties:
            print(f"\n  {prop.name}:")
            print(f"    Type: {prop.data_type}")
            print(f"    Skip vectorization: {getattr(prop, 'skip_vectorization', 'not set')}")
            # Check if it has any other special attributes
            for attr in dir(prop):
                if not attr.startswith('_') and attr not in ['name', 'data_type', 'skip_vectorization']:
                    val = getattr(prop, attr, None)
                    if val and not callable(val):
                        print(f"    {attr}: {val}")

finally:
    client.close()
    print("\nтЬУ Connection closed")
```
